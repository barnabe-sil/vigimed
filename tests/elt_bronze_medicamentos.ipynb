# ELT

```python
import pandas as pd, csv
from datetime import datetime
```

```python
layer ='01_bronze'
dataset = 'medicamentos'
ts = datetime.now().strftime("%Y_%m_%d_%H_%M_%S") #'2025_06_30_14_17_00'
input = r'C:\\Users\\silma\\Downloads\\' 
output = r"C:\\Users\\silma\\OneDrive\\Área de Trabalho\\Mestrado\\vigimed\\data\\"+layer+"\\"+dataset+"\\"+ts+".parquet"
print(output)
```

# Input

```python
df = pd.read_csv(input+'\\VigiMed_Medicamentos.csv', 
                 sep=';',
        encoding='latin-1',
        dtype=str,
        #engine='python',          # motor flexível
        quoting=csv.QUOTE_NONE,   # não trate aspas como delimitador
        #escapechar='\\',          # caso apareçam barras de escape
        on_bad_lines='error',     # queremos erro explícito se algo fugir do layout
        low_memory=False
                 )
```

```python
df.head()
```

# Output

```python
df.to_parquet(output, compression="snappy", index=False)
```
